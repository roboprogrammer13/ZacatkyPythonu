Odebere písně interpreta

```python
playlist_filtered = [song for song in playlist if song[0] != "Taylor Swift"]
print(playlist_filtered[0:5])
```

```python
for i in range(len(playlist) - 1, -1, -1):  # projdeme seznam pozpátku
    if playlist[i][0] == "Madonna":
        del playlist[i]

# kontrola výsledku
print("Počet písní po odebrání:", len(playlist))
print("Aktuální playlist:", playlist)
```

Zobrazit celkovou délku playlistu

```python
total_length = 0
for song in playlist:
    total_length += song[2]  # třetí prvek n-tice je délka v minutách

print(f"Celková délka playlistu: {total_length:.0f} minut")
```

Seznam unikátních interpretů

```python
unique_artists = set()
for song in playlist:
    unique_artists.add(song[0])  # první prvek n-tice je interpret

# výpis
print("Unikátní interpreti:")
for artist in unique_artists:
    print(artist, end=", ")
```

Seřadit podle délky

```python
# seřazení playlistu podle délky písní (od nejkratší po nejdelší)
playlist.sort(key=lambda song: song[2])

# výpis prvních 5 písní pro kontrolu
print("5 nejkratších písní v playlistu:")
for song in playlist[:5]:
    print(f"{song[0]} – {song[1]} ({song[2]} min)")
```
