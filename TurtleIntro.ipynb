### **Digitální** puls

```python
initializeTurtle(initial_speed=13)
bgcolor('black') # Základ pro vibe
colors = ['cyan', (0,200,255), 'white']

for i in range(120):
    color(colors[i % 3])    # Střídání 3 barev
    width(i % 9 + 1)       # Mění tloušťku čáry (Puls efekt)

    forward(i * 3)          # Stále delší čáry
    left(58)                # 30,60 90 x 58
```

### Kybernetická pavučina

```python
initializeTurtle(initial_speed=13)
bgcolor('black')
color('lime')      # Matrix zelená
width(1)

# Funkce pro nakreslení jednoho šestiúhelníku
def hexagon(velikost):
    for _ in range(6):
        forward(velikost)
        left(60)

# Hlavní smyčka
for i in range(30):
    hexagon(100)   # Nakresli tvar
    left(36)       # Pootoč se 12,24
```

### Příklad 1: Neonový Tunel (The Glitch Spiral)

Trik je v úhlu 89 nebo 91 stupňů. Kdyby to bylo 90, je to nuda čtverec. Ta jednička navíc vytváří "točivý" 3D efekt.
Změňte úhel v prvním příkladu z 91 na 121. Co se stane?"

```python
import random

initializeTurtle(initial_speed=10) # 13 je max rychlost
bgcolor('black') # Černé pozadí je základ pro neonový look
width(2) # Trochu tlustší čára je hezčí

# Paleta barev (Cyberpunk vibes)
colors = ['cyan', 'magenta', 'yellow', 'lime']

for i in range(150):
    color(colors[i % 4]) # Střídání 4 barev dokola
    forward(i * 2)       # Každý krok je o kousek delší
    left(91)             # Úhel 91° vytvoří ten spirálový efekt
```

### Příklad 2: Hypnotická Mandala (Symmetry Vibe)
Vypadá to jako složitý květ, ale je to jen spousta kruhů, které se mírně otáčejí. Tohle je velmi uklidňující na sledování.

```python
def muj_kruh(step):
    """Nakreslí 36-úhelník, který vypadá jako kruh."""
    for _ in range(36):
        forward(step)
        left(10)

def mandala(step=15, colors=None):
    """
    Nakreslí mandalu z 36-úhelníků.

    step   - délka strany každého 36-úhelníku
    colors - seznam barev; pokud None, použije defaultní barevný tuple
    """
    if colors is None:
        colors = [(255,0,0), (255,127,0), (255,255,0),
                  (0,255,0), (0,255,255), (255,0,255)]

    for i in range(36):
        color(colors[i % len(colors)])
        muj_kruh(step)
        left(10)  # pootočení mandaly

# -------------------------------
initializeTurtle(initial_speed=13, initial_window_size=(500,300))
speed(10)

# Příklad použití
mandala(step=15)
```

### Chaotická Hvězda (Random Vibe)
Zde využijeme náhodu (random). Žáci uvidí, že počítač může být nepředvídatelný. Pokaždé, když to spustí, vznikne jiný obrazec.

```python
initializeTurtle(initial_speed=10) # 13 je max rychlost
bgcolor('black') # Černé pozadí je základ pro neonový look
width(1) # Tenčí čára pro jemnější vzhled

colors = ['cyan', 'white', 'magenta']

# Cyklus běží 100x
for _ in range(100):
    color(random.choice(colors)) # Vybere náhodnou barvu
    length = random.randint(50, 180) # Náhodná délka paprsku

    forward(length)
    backward(length) # Vrátí se zpátky do středu
    right(random.randint(5, 30)) # Náhodně se otočí
```

```python
initializeTurtle(initial_speed=13)
bgcolor('black')
width(2)
colors = ['cyan', 'magenta']

for i in range(24): # 36 x 10, 12 x 30, 24 x 15
    color(colors[i % 2])

    # Nakreslí čtverec místo kruhu
    for _ in range(4):
        forward(100)
        left(90)

    left(15)
```

###Známé křivky- Archimedova spirála, Logaritmická spirála, Klotoida

```python
initializeTurtle(initial_speed=13,initial_window_size=(500,200))
bgcolor('black')
color('cyan')
width(2)

# Proměnná pro délku kroku
krok = 0.5

for i in range(300):
    forward(krok)
    left(8)        # Úhel je stále stejný
    krok += 0.05    # Délka kroku se pomalu lineárně zvětšuje

    # Efekt: změna barvy
    if i % 50 == 0:
        color('cyan' if i % 100 == 0 else 'magenta')
```

```python
initializeTurtle(initial_speed=13,initial_window_size=(300,300))
color('darkviolet')
width(2)

# Posuneme se trochu dolů, klotoida potřebuje místo
penup()
goto(50, 250)
setheading(0)
pendown()

uhel_otoceni = 0

for i in range(200):
    forward(9)              # Krok je konstantní!
    left(uhel_otoceni)      # Otáčíme se
    uhel_otoceni += 0.2     # Ale úhel se příště zvětší (zvyšujeme zakřivení)

    # Změna barvy pro hezčí efekt
    if i == 100:
        color('lime')
```

⭐ Lissajousova křivka —skládání dvou harmonických kmitání v kolmých směrech ([Youtube](https://youtu.be/ZNOTf-CjmQQ))

```python
from ColabTurtle.Turtle import *
import math

initializeTurtle(initial_speed=13,initial_window_size=(500,300))
penup()
hideturtle()

# Parametry křivky
A = 100   # amplituda X
B = 100   # amplituda Y
a = 3     # frekvence X
b = 2     # frekvence Y
delta = math.pi / 2  # fázový posun

# Vykreslení
for t in range(1000):
    x = A * math.sin(a * t * 0.01 + delta) + 200
    y = B * math.sin(b * t * 0.01) + 200
    goto(x, y)
    pendown()
```
