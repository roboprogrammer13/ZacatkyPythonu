### Prolomen√≠ PINU, break
Prostudujte p≈ô√≠kaz break na [w3schools](https://www.w3schools.com/python/ref_keyword_break.asp).


**Poƒçet mo≈æn√Ωch hesel = velikost abecedy ^ d√©lkahesla**

(^ znamen√° "na mocninu")

P≈ô√≠klad 1: Jednoduch√° hesla (jen mal√° p√≠smena), 26 p√≠smen (a-z)
- Heslo dlouh√© 4 znaky: 26 √ó 26 √ó 26 √ó 26 = **456 976 mo≈ænost√≠**
- Heslo dlouh√© 8 znak≈Ø: 26^8 = **208 miliard mo≈ænost√≠**

P≈ô√≠klad 2: Silnƒõj≈°√≠ hesla (mal√° + velk√° p√≠smena + ƒç√≠sla), 62 znak≈Ø (a-z, A-Z, 0-9)
- Heslo dlouh√© 4 znaky: 62^4 = **14 milion≈Ø mo≈ænost√≠**
- Heslo dlouh√© 8 znak≈Ø: 62^8 = **218 bilion≈Ø mo≈ænost√≠**


```python
from time import time
# Tajn√© heslo (6 ƒç√≠slic)
heslo = int(input("Zadej ƒç√≠slicov√© heslo: "))

# zaƒç√°tek mƒõ≈ôen√≠
start = time()

# Zkus√≠me v≈°echny kombinace od 0 do 999 999 999
for pokus in range(1000000000):
    if pokus == heslo:
        print("Heslo nalezeno:", pokus)

        # konec mƒõ≈ôen√≠
        end = time()
        print("Trvalo to:", round(end - start, 6), "sekund")
        break
```
