### Poƒçet ƒç√≠sel men≈°√≠ch ne≈æ k
üíéüíéNapi≈°te program, kter√Ω vr√°t√≠ poƒçet ƒç√≠sel men≈°√≠ch ne≈æ k v seznamu cisla.

Pou≈æijte List Comprehension `[novy for prvek in seznam if podminka]`

vytvo≈ôte seznam *kontrola* takov√Ω, ≈æe  `kontrola[i] = 1 `  kdy≈æ je `cisla[i] < k`.

````python
cisla = [1, 2, 3, 4, 5,0,-1]
k = 3
pocet = 0
for cislo in cisla:
    if cislo < k:
        pocet += 1
print(pocet)
````

````python
kontrola=[1 for cislo in cisla if cislo < k] # podm√≠nka urƒçuje, jestli se prvek v≈Øbec pou≈æije
print(kontrola)
sum(kontrola)
````

````python
kontrola=[1 if cislo<k else 0 for cislo in cisla]
print(kontrola)
````

````python
#row = [1, 2, 3, 4, 5, 6, 6, 8, 9]
row = list(range(1, 10))
if len(row) != 9:
    sudoku = False
else:
    sudoku = True
    for i in range(1, 10):
        if row.count(i) != 1:
            sudoku = False
            break
print(sudoku)
````

````python
def max_pair_sum(num_list):
    if len(num_list) < 2:
        return None  # nelze vytvo≈ôit dvojici

    max_sum = num_list[0] + num_list[1]
    for i in range(1, len(num_list) - 1):
        pair_sum = num_list[i] + num_list[i + 1]
        if pair_sum > max_sum:
            max_sum = pair_sum
    return max_sum
print(max_pair_sum([1, 8, 7, 3, 5, 2]))
````

### Jeden v√Ωskyt
üíéüíéüíéNapi≈°te funkci unique(mylist), kter√° vr√°t√≠ seznam obsahuj√≠c√≠ ka≈æd√Ω prvek ze seznamu mylist pr√°vƒõ jedenkr√°t (v po≈ôad√≠ prvn√≠ch v√Ωskyt≈Ø v p≈Øvodn√≠m seznamu).

````python
# neidiomaticky
def unique(mylist):
    result = []
    for x in mylist:
        if x not in result:
            result.append(x)
    return result

print(unique([1, 2, 3, 4, 3, 2, 1]))
````

````python
# list comprehension
def unique(mylist):
    result = []
    [result.append(x) for x in mylist if x not in result]
    return result
print(unique([1, 2, 3, 4, 3, 2, 1]))
````

````python
# set
def unique(mylist):
    list_set = set(mylist)
    unique_list = (list(list_set))
    return unique_list
print(unique([1, 2, 3, 4, 3, 2, 1]))
````

### Da≈à

üíéüíéNapi≈°te funkci five_multiples(num_list), kter√° vezme seznam ƒç√≠sel num_list a vr√°t√≠ seznam tƒõch ƒç√≠sel ze seznamu, kter√° jsou dƒõliteln√° pƒõti (v p≈Øvodn√≠m po≈ôad√≠).

````python
def five_multiples(num_list):
    nasobky = []
    for cislo in num_list:
        if cislo % 5 == 0:
            nasobky.append(cislo)
    return nasobky

print(five_multiples([3, 5, 25, 4, 7, 49, 35, 12]))
````

````python
def five_multiples(num_list):
    return [num for num in num_list if num % 5 == 0]

print(five_multiples([3, 5, 25, 4, 7, 49, 35, 12]))
````

### P√°rov√© minimum
üíéNapi≈°te program kter√Ω vr√°t√≠ nejmen≈°√≠ rozd√≠l dvou ƒç√≠sel ze seznamu numbers.

````python
numbers = [3, 8, 15, 90, 18]

if len(numbers) < 2:
    print("Nelze porovnat m√©nƒõ ne≈æ dvƒõ ƒç√≠sla")
else:
    # se≈ôad√≠me ƒç√≠sla
    numbers_sorted = sorted(numbers)

    # seznam rozd√≠l≈Ø mezi sousedn√≠mi prvky
    differences = [numbers_sorted[i+1] - numbers_sorted[i] for i in range(len(numbers_sorted)-1)]

    # nejmen≈°√≠ rozd√≠l
    min_diff = min(differences)

    print("Nejmen≈°√≠ rozd√≠l je:", min_diff)
````

````python
def minimum_difference(numbers):
    if len(numbers) < 2:
        return None  # nelze porovnat m√©nƒõ ne≈æ dvƒõ ƒç√≠sla

    # Se≈ôad√≠me ƒç√≠sla, abychom mohli porovnat jen sousedn√≠ hodnoty
    numbers_sorted = sorted(numbers)
    min_diff = abs(numbers_sorted[1] - numbers_sorted[0])

    for i in range(1, len(numbers_sorted) - 1):
        diff = abs(numbers_sorted[i + 1] - numbers_sorted[i])
        if diff < min_diff:
            min_diff = diff
    return min_diff

print(minimum_difference([3, 8, 15, 90, 18]))
````

### Operace na √∫ƒçtu
üíéNapi≈°te funkci balance_check(operations), kter√° dostane seznam operations operac√≠ na √∫ƒçtu (p≈ôips√°n√≠ penƒõz / v√Ωbƒõr) a zkontroluje, zda je suma na √∫ƒçtu po celou dobu nez√°porn√°.

````python
def balance_check(operations):
    stav =  0
    for transakce in operations:
        stav  = stav + transakce
        if stav < 0:
            return False
    return True

print(balance_check([100, -50, 30, -60, 400, 200, -550]))
````

### Mo≈æn√© souƒçty
üíéNapi≈°te funkci pairs_sum(numbers, x), kter√° vyp√≠≈°e v≈°echny zp≈Øsoby, jak lze ƒç√≠slo x vyj√°d≈ôit jako souƒçet dvou ƒç√≠sel ze seznamu numbers. M≈Ø≈æete p≈ôedpokl√°dat, ≈æe seznam numbers obsahuje kladn√° cel√° ƒç√≠sla a ≈æe se v nƒõm ≈æ√°dn√© ƒç√≠slo neopakuje. V√Ωpis mus√≠ b√Ωt se≈ôazen√Ω podle prvn√≠ho sƒç√≠tance.

````python
def pairs_sum(numbers, x):
    numbers = sorted(numbers)  # se≈ôad√≠me podle velikosti
    for i in range(len(numbers)):
        for j in range(i + 1, len(numbers)):
            if numbers[i] + numbers[j] == x:
                print(numbers[i], "+", numbers[j], "=", x)

pairs_sum([10, 13, 5, 9, 12, 8, 3], 15)
````

### Monotonn√≠ posloupnost
üíéNapi≈°te funkci monotonic(numbers), kter√° vr√°t√≠ True, pokud je seznam ƒç√≠sel numbers rostouc√≠ nebo klesaj√≠c√≠.

````python
def monotonic(numbers):
    return numbers == sorted(numbers) or numbers == sorted(numbers, reverse=True)

print(monotonic([1, 12, 30, 35]))
````

````python
def monotonic(numbers):
    increasing = all(numbers[i] <= numbers[i + 1] for i in range(len(numbers) - 1))
    decreasing = all(numbers[i] >= numbers[i + 1] for i in range(len(numbers) - 1))
    return increasing or decreasing

print(monotonic([1, 12, 30, 35]))
````
