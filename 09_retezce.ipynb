üíéüíéNapi≈°te funkci join(text1, text2), kter√° vr√°t√≠ spojen√≠ tƒõchto dvou ≈ôetƒõzc≈Ø, p≈ôiƒçem≈æ ten krat≈°√≠ z nich bude prvn√≠.

```python
def join(text1, text2):
    if len(text1) < len(text2):
        return text1+text2
    else:
        return text2+text1

join("Anaconda", "Python")
```

üíéüíéüíéNapi≈°te funkci duplication(text), kter√° pro zadan√Ω ≈ôetƒõzec vr√°t√≠ ≈ôetƒõzec, ve kter√©m je ka≈æd√© p√≠smeno duplikov√°no.

```python
def duplication(text):
    duplicitni = ""
    for znak in text:
        duplicitni += znak * 2  # Duplikuje ka≈æd√© p√≠smeno
    return duplicitni

duplication("Python")
```

üíéüíéüíéNapi≈°te funkci palindrom(text), kter√° otestuje, zda je zadan√Ω ≈ôetƒõzec text palindromem (ƒçte se stejnƒõ od zaƒç√°tku jako od konce).

```python
def palindrom(text):
    n=len(text)
    rozhodnuti = True
    for i in range (n//2):
        rozhodnuti = rozhodnuti and text[i]==text[n-i-1]
    return rozhodnuti

palindrom("jeblbej")
palindrom("nodevillivedon") # No devil lived on. ‚Äû≈Ω√°dn√Ω ƒè√°bel ne≈æil d√°l"
```

üíéüíéüíéNapi≈°te funkci helke(text), kter√° vyp√≠≈°e zadan√Ω text, ve kter√©m ov≈°em v≈°echny samohl√°sky nahrad√≠ za p√≠smeno e. M≈Ø≈æete p≈ôedpokl√°dat, ≈æe text obsahuje pouze mal√° p√≠smena anglick√© abecedy.

üíéüíéüíéNapi≈°te funkci frame(text, symbol), kter√° vyp√≠≈°e text v r√°meƒçku tvo≈ôen√©m znakem symbol (p≈ôedpokl√°dejte, ≈æe symbol je ≈ôetƒõzec tvo≈ôen√Ω pr√°vƒõ jedn√≠m znakem).

```python
def frame(text, symbol):
    n=len(text)
    print(symbol*(n+2))
    print(symbol+text+symbol)
    print(symbol*(n+2))

frame("Ahoj", "*")
```

## Tƒõ≈æ≈°√≠
üíéüíéüíéNapi≈°te funkci censorship(text), kter√° vr√°t√≠ zadan√Ω ≈ôetƒõzec text, ve kter√©m nahrad√≠ ka≈æd√© druh√© p√≠smeno za X.

```python
text="J√≠zdn√≠ pruh, ve kter√©m stoj√≠te, se pohybuje v≈ædy nejpomaleji."
i = 1
for znak in text:
    if i%2==0:
        print("X", end="")
    else:
        print(znak, end="")
    i+=1
```

```python
def censorship(text):
    cenzura = ""
    for i, znak in enumerate(text):
        if (i + 1) % 2 == 0:  # Kontrolujeme, zda je index (pozice) lich√Ω
            cenzura += 'X'
        else:
            cenzura += znak
    return cenzura

censorship("J√≠zdn√≠ pruh, ve kter√©m stoj√≠te, se pohybuje v≈ædy nejpomaleji.")
```

Pou≈æijeme list [comprehension](https://www.w3schools.com/python/python_lists_comprehension.asp#gsc.tab=0) a nakonec spoj√≠me seznam do ≈ôetƒõzce pomoc√≠` "".join()`.
Pro z√≠sk√°n√≠ sƒç√≠tac√≠ho indexu pou≈æijeme` enumerate`.

Funkce `enumerate()` se pou≈æ√≠v√°, kdy≈æ chceme iterovat p≈ôes seznam (nebo jin√Ω iterable) a z√°rove≈à pot≈ôebuje≈° index ka≈æd√©ho prvku.

`for index, hodnota in enumerate(iterable, start=0):`


```python
text="J√≠zdn√≠ pruh, ve kter√©m stoj√≠te, se pohybuje v≈ædy nejpomaleji."
cenzurovane = ["X" if (i + 1) % 2 == 0 else znak for i, znak in enumerate(text)]
print(cenzurovane[2:20:2])
print("".join(cenzurovane))
```

```python
def censorship(text):
    return "".join(
        'X' if (i + 1) % 2 == 0 else znak
        for i, znak in enumerate(text)
    )

print(censorship("J√≠zdn√≠ pruh, ve kter√©m stoj√≠te, se pohybuje v≈ædy nejpomaleji."))
```

üíéüíéüíéNapi≈°te funkci string_intersection(left, right), kter√° pro dva zadan√© ≈ôetƒõzce left a right vyp√≠≈°e p√≠smena, kter√° jsou v obou ≈ôetƒõzc√≠ch na stejn√© pozici.

```python
def string_intersection(left, right):
    if len(left) > len(right):
        n=len(right)
    else:
        n=len(left)
    for i in range(n):
        if left[i]==right[i]:
            print(left[i])

string_intersection("Kobyla", "Mohyla")
```
