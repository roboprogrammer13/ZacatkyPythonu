```python
trams = [
    ("Tatra T1", 60, 1951),
    ("Tatra T2", 65, 1955),
    ("Tatra T3", 65, 1960),
    ("Tatra T3SUCS", 65, 1976),
    ("Tatra T3R.P", 65, 2000),
    ("Tatra T3M", 65, 1976),
    ("Tatra T6A5", 65, 1991),
    ("Tatra KT4D (ex-DDR)", 65, 1976),
    ("Tatra K2", 65, 1966),
    ("Tatra K2S", 65, 2003),
    ("Tatra KT8D5", 65, 1984),
    ("Tatra KT8D5.RN2P", 65, 2005),
    ("Škoda 14T", 60, 2005),
    ("Škoda 15T", 60, 2009),
    ("RT6N1", 60, 1993),
    ("VarioLF", 65, 2003),
    ("VarioLF2", 65, 2009),
    ("Pragoimex EVO1", 50, 2015),
    ("Pragoimex EVO2", 60, 2015)
]
```

```python
def treti_polozka(x):
    return x[2]
```

```python
trams_sorted = sorted(trams, key=treti_polozka)
#trams_sorted = sorted(trams, key=lambda x: x[2])
for t in trams_sorted:
    print(t)
```

Filtrování podle roku

```python
modern = [t for t in trams if t[2] >= 2000]
print(modern)
```

Průměrná rychlost všech tramvají

```python
avg_speed = sum(t[1] for t in trams) / len(trams)
print("Průměrná rychlost:", avg_speed)
```

Histogram podle roku

```python
import matplotlib.pyplot as plt

roky = [rok for (_, _, rok) in trams]

plt.hist(roky, bins=10)   # lze změnit počet košů
plt.xlabel("Rok výroby")
plt.ylabel("Počet typů")
plt.title("Histogram tramvají podle roku výroby")
plt.show()
```
